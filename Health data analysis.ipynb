```python
df1_cleaned = df1.dropna(subset=['value'])

# Filter rows where any word contains "health"
filtered_rows = df1_cleaned[df1_cleaned['value'].str.split().apply(lambda words: any("health" in word for word in words))].copy()

# Function to extract words before "health"
def extract_words_before_health(value):
    words = value.lower().split()
    health_index = None
    for i, word in enumerate(words):
        if "health" in word:
            health_index = i
            break
    if health_index is not None:
        extracted_words = ' '.join(words[:health_index])
        return extracted_words
    else:
        return None

# Apply the function using .loc to create a new column
filtered_rows.loc[:, 'extracted_words'] = filtered_rows['value'].apply(extract_words_before_health)


filtered_rows
```

```python
class_counts = df3['value'].explode().value_counts()

# Convert class_counts Series to a DataFrame and reset the index
class_counts_df = class_counts.reset_index()

# Rename the columns
class_counts_df.columns = ['class', 'frequency']

# Sort the DataFrame by frequency in descending order
class_counts_df = class_counts_df.sort_values(by='frequency', ascending=False)

# Reset the index again
class_counts_df.reset_index(drop=True, inplace=True)
ds=class_counts_df.head(10)
print(ds)
plt.figure(figsize=(20, 16))
plt.pie(ds['frequency'], labels=ds['class'], autopct='%1.1f%%', startangle=140)
plt.title('Top 10 Most Important categories chosen by People')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
categories = [
    "sexually_active",
    "no_sickness",
    "looking_good",
    "mental_sharpness",
    "health_wellbeing",
    "fit_and_strong",
    "living_long",
    "lose_weight"
]

# Create new columns for each category
for category in categories:
    df3[category] = df3['value'].apply(lambda x: category if category in x else None)
    
df3
```

```python
value_counts_per_category = {}
for category in categories:
    value_counts_per_category[category] = df3[category].value_counts(dropna=False)

# Display value counts
for category, value_counts in value_counts_per_category.items():
    print(f"Value counts for '{category}':")
    print(value_counts)
    print()
```

```python
accumulated_counts = {}

# Accumulate the counts for the specific category value for each category
for category in categories:
    category_value_counts = value_counts_per_category[category]
    category_count = category_value_counts.get(category, 0)
    accumulated_counts[category] = category_count

# Plotting the accumulated counts in a single bar chart
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
plt.bar(accumulated_counts.keys(), accumulated_counts.values(), color='steelblue')
plt.title("Highest Important factor Values")
plt.xlabel("Category")
plt.ylabel("Number of people that chose")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```
